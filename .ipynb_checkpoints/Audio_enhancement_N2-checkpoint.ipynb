# Notebook 2: Audio Enhancement Algorithm and Evaluation
## Project: Team Communication Processing and Analysis
### GSoC 2026 — HumanAI / TRIP Lab Screening Task

This notebook implements and evaluates audio enhancement methods to improve
clarity of team communication recordings for downstream transcription.

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import scipy.signal as signal
import warnings
warnings.filterwarnings('ignore')

print("Libraries loaded.")
```

```python
input_path = "sample_input.wav"
y, sr = librosa.load(input_path, sr=None)
duration = librosa.get_duration(y=y, sr=sr)

print(f"Loaded: {input_path}")
print(f"Sample rate: {sr} Hz | Duration: {duration:.2f}s")
```

```python
# Simulate environment noise (fan, HVAC, equipment hum)
# This makes the task realistic — real lab recordings have this
np.random.seed(42)
noise_level = 0.015
noise = np.random.normal(0, noise_level, len(y))

# Add low-frequency hum (60Hz — common in labs)
t = np.linspace(0, duration, len(y))
hum = 0.008 * np.sin(2 * np.pi * 60 * t)

y_noisy = y + noise + hum
y_noisy = np.clip(y_noisy, -1.0, 1.0)

sf.write("sample_noisy.wav", y_noisy, sr)
print("Noisy audio saved: sample_noisy.wav")
```

```python
def spectral_subtraction(y, sr, noise_frames=10, alpha=2.0):
    """
    Spectral subtraction noise reduction.
    Estimates noise from initial frames and subtracts from signal.
    alpha controls aggressiveness of subtraction.
    """
    # STFT
    D = librosa.stft(y)
    magnitude = np.abs(D)
    phase = np.angle(D)
    
    # Estimate noise from first noise_frames frames
    noise_mag = np.mean(magnitude[:, :noise_frames], axis=1, keepdims=True)
    
    # Subtract noise
    magnitude_denoised = magnitude - alpha * noise_mag
    magnitude_denoised = np.maximum(magnitude_denoised, 0.1 * magnitude)
    
    # Reconstruct
    D_denoised = magnitude_denoised * np.exp(1j * phase)
    y_denoised = librosa.istft(D_denoised, length=len(y))
    
    return y_denoised

y_ss = spectral_subtraction(y_noisy, sr)
sf.write("enhanced_spectral_subtraction.wav", y_ss, sr)
print("Method 1 complete: Spectral Subtraction")
```

```python
def wiener_filter(y, sr, noise_frames=10):
    """
    Wiener filter: estimates optimal filter in frequency domain
    to minimize mean square error between clean and noisy signal.
    """
    D = librosa.stft(y)
    power = np.abs(D)**2
    
    # Estimate noise power
    noise_power = np.mean(power[:, :noise_frames], axis=1, keepdims=True)
    
    # Wiener gain
    gain = power / (power + noise_power + 1e-10)
    
    D_filtered = gain * D
    y_wiener = librosa.istft(D_filtered, length=len(y))
    
    return y_wiener

y_wiener = wiener_filter(y_noisy, sr)
sf.write("enhanced_wiener.wav", y_wiener, sr)
print("Method 2 complete: Wiener Filter")
```

```python
def highpass_normalize(y, sr, cutoff=80):
    """
    High-pass filter to remove low-frequency hum (HVAC, equipment),
    followed by RMS normalization for consistent loudness.
    """
    # High-pass Butterworth filter
    nyq = sr / 2
    b, a = signal.butter(4, cutoff / nyq, btype='high')
    y_hp = signal.filtfilt(b, a, y)
    
    # RMS normalization
    rms = np.sqrt(np.mean(y_hp**2))
    y_norm = y_hp / (rms + 1e-10) * 0.1
    
    return y_norm

y_hp = highpass_normalize(y_noisy, sr)
sf.write("enhanced_highpass.wav", y_hp, sr)
print("Method 3 complete: High-pass + Normalization")
```

```python
def full_enhancement_pipeline(y, sr):
    """
    Full pipeline: High-pass filter → Spectral Subtraction → Wiener Filter
    Combines all three methods for maximum clarity improvement.
    """
    y_step1 = highpass_normalize(y, sr)
    y_step2 = spectral_subtraction(y_step1, sr)
    y_step3 = wiener_filter(y_step2, sr)
    return y_step3

y_enhanced = full_enhancement_pipeline(y_noisy, sr)
sf.write("sample_output.wav", y_enhanced, sr)
print("Full pipeline complete: sample_output.wav saved")
```

```python
fig, axes = plt.subplots(4, 1, figsize=(14, 14))

signals = [
    (y_noisy,    "Original Noisy Audio",             'tomato'),
    (y_ss,       "After Spectral Subtraction",        'steelblue'),
    (y_wiener,   "After Wiener Filter",               'seagreen'),
    (y_enhanced, "Final Enhanced (Full Pipeline)",    'purple'),
]

for ax, (sig, title, color) in zip(axes, signals):
    librosa.display.waveshow(sig, sr=sr, ax=ax, color=color)
    ax.set_title(title, fontsize=12)
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("Amplitude")

plt.tight_layout()
plt.savefig("enhancement_comparison_waveforms.png", dpi=150, bbox_inches='tight')
plt.show()
print("Comparison plot saved.")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

pairs = [
    (y_noisy,    "Noisy Input"),
    (y_ss,       "Spectral Subtraction"),
    (y_wiener,   "Wiener Filter"),
    (y_enhanced, "Full Pipeline (Final)"),
]

for ax, (sig, title) in zip(axes.flatten(), pairs):
    D = librosa.amplitude_to_db(np.abs(librosa.stft(sig)), ref=np.max)
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='hz', ax=ax, cmap='magma')
    ax.set_title(title, fontsize=11)

plt.suptitle("Spectrogram Comparison — Enhancement Methods", fontsize=14, y=1.01)
plt.tight_layout()
plt.savefig("enhancement_spectrogram_comparison.png", dpi=150, bbox_inches='tight')
plt.show()
print("Spectrogram comparison saved.")
```

```python
def compute_snr(clean, noisy):
    noise = clean - noisy[:len(clean)]
    snr = 10 * np.log10(np.mean(clean**2) / (np.mean(noise**2) + 1e-10))
    return snr

def compute_rms(y):
    return np.sqrt(np.mean(y**2))

def spectral_flatness(y):
    D = np.abs(librosa.stft(y))
    geo_mean = np.exp(np.mean(np.log(D + 1e-10)))
    arith_mean = np.mean(D)
    return geo_mean / (arith_mean + 1e-10)

ref = y[:min(len(y), len(y_noisy))]

print("=" * 55)
print("  AUDIO ENHANCEMENT EVALUATION METRICS")
print("=" * 55)
results = {
    "Noisy Input":           y_noisy,
    "Spectral Subtraction":  y_ss,
    "Wiener Filter":         y_wiener,
    "Full Pipeline":         y_enhanced,
}
for name, sig in results.items():
    snr = compute_snr(ref, sig)
    rms = compute_rms(sig)
    sf_val = spectral_flatness(sig)
    print(f"\n  [{name}]")
    print(f"    SNR              : {snr:+.2f} dB")
    print(f"    RMS Energy       : {rms:.5f}")
    print(f"    Spectral Flatness: {sf_val:.5f}")
print("=" * 55)
print("\nHigher SNR = better noise reduction.")
print("Lower spectral flatness = more tonal (speech-like) signal.")
```

## Conclusion

In this experiment, I implemented and tested **three different audio enhancement methods** to improve speech clarity.

| Method                   | Why I Used It                                               | What It Does                                   |
| ------------------------ | ----------------------------------------------------------- | ---------------------------------------------- |
| **High-pass Filter**     | To remove low-frequency hum (like 60Hz lab equipment noise) | Reduces HVAC and electrical hum noise          |
| **Spectral Subtraction** | To remove constant background noise                         | Makes the speech spectrum cleaner              |
| **Wiener Filter**        | For smarter noise reduction in frequency domain             | Produces the smoothest and most natural output |

Finally, I combined all three methods into a **single pipeline** to get the best possible transcription clarity.

---

### Key Findings

* Signal-to-Noise Ratio (SNR) improved with all methods.
* **Spectral subtraction** worked best for constant (stationary) background noise.
* **Wiener filter** preserved speech details and harmonics better than the others.
* The **combined pipeline** gave the most balanced and clean result, so it is recommended for real TRIP Lab recordings.
